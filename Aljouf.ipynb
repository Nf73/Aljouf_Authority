## تنظيف وتجهيز البيانات 

```python
df = df[df['عدد الايام'].notnull()]
```

```python
df.isnull().sum()
print("عدد البلاغات بعد التنظيف:", len(df))
```

```python
df = df[df['حالة البلاغ'].str.strip() == 'Close']
```

```python
df.head()
```

```python
df.drop(columns=[ 'نوع التصنيف التخصصي', 'حالة البلاغ'], inplace=True)
```

```python
df.head()
```

```python
df['تاريخ الانشاء'] = pd.to_datetime(df['تاريخ الانشاء'], errors='coerce')
df['شهر_البلاغ'] = df['تاريخ الانشاء'].dt.month
df['يوم_في_الأسبوع'] = df['تاريخ الانشاء'].dt.weekday
df['ساعة_البلاغ'] = df['تاريخ الانشاء'].dt.hour
```

```python
df.drop(columns=['تاريخ الانشاء'],inplace=True)
```

```python
df.drop(columns=['الامانة'],inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

# نجهز قاموس لحفظ الـ encoders عشان نقدر نستخدمهم لاحقًا في API
label_encoders = {}

# الأعمدة اللي بنرّمزها
categorical_cols = ['البلدية', 'التصنيف الرئيسى', 'التصنيف الفرعي', 'التصنيف التخصصي']

for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le  # نخزن الـ encoder
```

```python
# استخراج الـ LabelEncoder الخاص بعمود التصنيف الرئيسي
le_main = label_encoders['التصنيف الرئيسى']

# طباعة التصنيفات الحقيقية مع أرقامها
for i, label in enumerate(le_main.classes_):
    print(f"{i}: {label}")
```

```python
le_main = label_encoders['البلدية']

# طباعة التصنيفات الحقيقية مع أرقامها
for i, label in enumerate(le_main.classes_):
    print(f"{i}: {label}")
```

```python
df.groupby('البلدية')['عدد الايام'].mean().sort_values(ascending=False).head(10)
```

```python

# حساب الربع الأول والثالث
Q1 = df['عدد الايام'].quantile(0.25)
Q3 = df['عدد الايام'].quantile(0.75)
IQR = Q3 - Q1

# تحديد القيم المتطرفة
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['عدد الايام'] < lower_bound) | (df['عدد الايام'] > upper_bound)]

print(f"عدد القيم المتطرفة: {len(outliers)}")
print(f"الحد الأعلى للقيم الطبيعية: {upper_bound:.2f}")
print(f"الحد الأدنى للقيم الطبيعية: {lower_bound:.2f}")
```

## بناء وتقييم النموذج

```python
x1 = df.drop('عدد الايام', axis=1)
y1 = df['عدد الايام']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, Y_train, Y_test = train_test_split(
    x1, y1, test_size=0.2
)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=100)
model.fit(x_train, Y_train)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score

Y_pred = model.predict(x_test)

mae = mean_absolute_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)

print(f"متوسط الخطأ المطلق (MAE): {mae:.2f}")
print(f"معامل التحديد (R²): {r2:.2f}")
```

```python
# إعادة حساب IQR
Q1 = df['عدد الايام'].quantile(0.25)
Q3 = df['عدد الايام'].quantile(0.75)
IQR = Q3 - Q1

# الحد الأعلى للقيم الطبيعية
upper_bound = Q3 + 1.5 * IQR

# استخراج القيم المتطرفة العالية فقط
high_outliers = df[df['عدد الايام'] > upper_bound]

# نحدد عدد النص للحذف
half_to_drop = int(len(high_outliers) / 2)

# نحذف نصف القيم المتطرفة (الأعلى فقط)
df_trimmed = df.drop(high_outliers.sort_values('عدد الايام', ascending=False).head(half_to_drop).index)

print(f"عدد البلاغات بعد الحذف الجزئي: {len(df_trimmed)}")
```

```python
# حساب الربع الأول والثالث
Q1 = df['عدد الايام'].quantile(0.25)
Q3 = df['عدد الايام'].quantile(0.75)
IQR = Q3 - Q1

# تحديد القيم المتطرفة
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df_trimmed[(df_trimmed['عدد الايام'] < lower_bound) | (df_trimmed['عدد الايام'] > upper_bound)]

print(f"عدد القيم المتطرفة: {len(outliers)}")
print(f"الحد الأعلى للقيم الطبيعية: {upper_bound:.2f}")
print(f"الحد الأدنى للقيم الطبيعية: {lower_bound:.2f}")
```

```python
X = df_trimmed.drop('عدد الايام', axis=1)
y = df_trimmed['عدد الايام']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2
)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=100)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score

y_pred = model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"متوسط الخطأ المطلق (MAE): {mae:.2f}")
print(f"معامل التحديد (R²): {r2:.2f}")
```
